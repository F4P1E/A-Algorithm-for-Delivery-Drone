#### 2. Heuristic Functions:
    Various distance metrics can be employed to create a heuristic function, such as Euclidean distance, Manhattan distance, Minkowski distance, or Chebyshev distance.
##### Manhattan distance:
    The Manhattan heuristic uses the Manhattan distance to estimate the remaining distance from the current position to the goal. This distance is calculated as the sum of the absolute differences between their coordinates.

##### Formula for this: $D_M(p, q) = sum(|p_x - q_x|, |p_y - q_y)$
    Components of this formula:
    - $D_M$: The Manhattan distance between points $p$ and $q$.
    - $p_x$: The x-coordinate of point $p$.
    - $p_y$: The y-coordinate of point $p$.
    - $q_x$: The x-coordinate of point $q$.
    - $q_y$: The y-coordinate of point $q$.
##### Use Case:
    Grid maps where movement is restricted to horizozntal and vertical directions only.
##### Benefits:
    Simple to implement, as the formula involves basic mathematical operations (addition, subtraction, and absolute value).
##### Drawbacks:
    1. It does not consider diagonal or other types of movement, which may result in inefficient pathfinding when diagonal moves are possible.
    2. It is not suitable for non-grid environments where movements are continuous rather than discrete.

```python
"""
current_location :The current location of the drone
goal :The destination the drone is trying to reach
result :The value that the heuristic function returns
0 and 1 are indexes of each coordinate of the location
"""
def heuristicManhattan(current_location, goal):
    # Calculate the Manhattan distance between the current_location and the goal.
    # This is done by summing the absolute differences of their respective x and y coordinates.
    result = abs(goal[0] - current_location[0]) + abs(goal[1] - current_location[1])

    # Return the calculated Manhattan distance as the result.
    return result

"""
`abs()`: A Python function that returns the absolute value of a number.
"""
```

##### Euclidean Distance
    This distance is determined by taking the square root of the sum of the squared differences between each corresponding coordinate of two points
##### Formula for this: $D_E(p, q) = \sqrt{(p_x - q_x) ^ 2 + (p_y - q_y) ^2}$
    Components of this formula
    - $D_E$: The Euclidean distance between the two points $p$ and $q$.
    - $p_x$: The x-coordinate of point $p$
    - $p_y$: The y-coordinate of point $p$
    - $q_x$: The x-coordinate of point $q$
    - $q_y$: The y-coordinate of point $q$
##### Use case:
    Euclidean distance represents the shortest path between two points in Euclidean space, which is the type of space we experience in the physical world. This makes it ideal for applications that require realistic distance calculations, such as robotics or computer graphics, where movement can occur in any direction, including diagonally.
##### Benefits
    Euclidean distance enables smooth and uninterrupted movements, making it suitable for robotic applications where fluid motion is desired.
##### Drawbacks
    1. The use of the square root operation in its formula adds computational complexity compared to other distance measures, potentially leading to higher resource consumption
    2. It is less suitable for grid-based scenarios where diagonal movement is restricted or impossible.

```python
"""
current_location :The current location of the drone
goal :The destination the drone is trying to reach
result :The value that the heuristic function returns
0 and 1 are indexes of each coordinate of the location
"""
def heuristicEuclidean(current_location, goal):
    # Calculate the Manhattan distance between the current_location and the goal.
    # This is done by summing the absolute differences of their respective x and y coordinates.
    result = ((goal[0] - current_location[0]) ** 2 + (goal[1] - current_location[1]) ** 2) ** (1 / 2)

    # Return the computed Euclidean distance as the heuristic value
    return result
```

##### Chebyshev distance:
    This distance is defined as the maximum of the absolute differences between the coordinates of two points.
##### Formula for this: $D_C(p, q) = max(|p_x - q_x|, |p_y - q_y|)$
    Components for this formula:
    - $D_C$: The Chebyshev distance between the points $p$ and $q$.
    - $p_x$: The x-coordinate of point $p$
    - $p_y$: The y-coordinate of point $p$
    - $q_x$: The x-coordinate of point $q$
    - $q_y$: The y-coordinate of point $q$
##### Use Case
    Chebyshev distance is ideal for grid-based maps where movement can occur horizontally, vertically, or diagonally. This makes it particularly useful in certain scenarios, such as video games and simulations.
##### Benefits:
    1. The Chebyshev distance is straightforward to implement, requiring only the use of the `max()` function and basic subtraction.
    2. It generally provides accurate results for relevant scenarios.
##### Drawbacks:
    1. Like the Manhattan distance, Chebyshev distance is not suitable for calculating distances in non-grid environments.
    2. Similar to Euclidean distance, it is also ineffective in environments where diagonal movements are not permitted.

```python
"""
current_location :The current location of the drone
goal :The destination the drone is trying to reach
result :The value that the heuristic function returns
0 and 1 are indexes of each coordinate of the location
"""
def heuristicChebyshev(current_location, goal):
    # Calculate the Manhattan distance between the current_location and the goal.
    # This is done by summing the absolute differences of their respective x and y coordinates.
    result = max(goal[0] - current_location[0], goal[1] - current_location[1])

    # Return the computed Chebyshev distance as the heuristic value
    return result

"""
`max()`: A Python function that returns the largest value among the given arguments.
"""
```

#### 3. A* Implementation:
    The A* search algorithm is implemented in Python as shown below, with each step explained through comments:

```python
# Import the heap queue (priority queue) data structure to use for the open list, which is explained further below.
import heapq
'''
Define a function named `a_star` which implements the A* search algorithm.
It takes three arguments: The city map, the starting location of the drone and a tuple of the delivery locations.
'''
def a_star(city_map, starting_location, delivery_location, heuristic):
    """
    The open list is a list of possible tiles that the drone may move to.
    The closed set is a set of tiles that the drones have already passed on its way to the destination.
    """
    #Initialize the open list as an empty priority queue:
    open_list = []
    #Initialize the closed list as an empty list:
    closed_list = []
    """
    The f_score represents the estimated total distance from starting location to the goal.
    It is the sum of the g_score (actual distance from the start to the current location)
    and the heuristic estimate from the current location to the delivery location.
    The f_score of the starting location is just the heuristic value since its g_score is 0.
    We push the f_score of the starting location into the open list:
    """
    heapq.heappush(open_list, (heuristic(starting_location, delivery_location), starting_location))
    # The `history` dictionary stores the path taken by the drone,
    # with each key being a location and the value being the location from which the drone arrived there.
    history = {}
    """
    `g_score`: A dictionary that holds `g_score`.
    It starts with an entry `starting_location: 0`
    Which indicates that the distance between the starting location and itself is 0.
    """
    g_score = {starting_location: 0}
    # Initialize the f_score dictionary with the starting location's f_score (heuristic value).
    f_score = {starting_location: heuristic(starting_location, delivery_location)}

    # Continue searching as long as there are locations in the open list:
    while open_list:
        # Extract the location with lowest f_score from the open list using heapq.heappop().
        _, current_location = heapq.heappop(open_list)
        # If the current location is the delivery location, the path has been found:
        if current_location == delivery_location:
            # Create an empty list `path` to store the reconstructed path:
            path = []
            # Trace back from the delivery location to the starting location using the `history` dictionary:
            while current_location in history:
                # Add the current location to the path:
                path.append(current_location)
                # Move to the previous location in the path:
                current_location = history[current_location]
            #Add the starting location to the path:
            path.append(starting_location)
            # Reverse the path since it was constructed from destination to start:
            path.reverse()
            # Store the results, including the path, the final drone location, the open list, and the closed list as a tuple:
            results = (path, list(delivery_location), open_list, closed_list)
            # Return the results:
            return results
        # If the current location is not the destination, add it to the closed list:
        closed_list.append(current_location)
        
        # Define the possible movements of the drone as the list of neighbor offsets:
        neighbours = [(-1, 0), (0, -1), (0, 1), (1, 0), (1, -1), (-1, -1), (-1, 1), (1, 1)]
        """
        Loop through each neighbour direction (dx, dy) to evaluate the neighboring tiles:
        """
        for dx, dy in neighbours:
            # Calculate the neighbour's coordinates by offsetting the current location:
            neighbour = (current_location[0] + dx, current_location[1] + dy)
            """
            Check if the neighbour is a valid tile:
            - 0 <= neighbour[0] < len(city_map): Ensure the neighbour is within the map's x-bounds.
            - 0 <= neighbour[1] < len(city_map[0]): Ensure the neighbour is within the map's y-bounds.
            - city_map[neighbour[0]][neighbour[1]] == 0: Ensure the neighbour is not an obstacle.
            - neighbour not in closed_list: Ensure the neighbour hasn't already been visited.
            """
            if 0 <= neighbour[0] < len(city_map) and 0 <= neighbour[1] < len(city_map[0]) and city_map[neighbour[0]][neighbour[1]] == 0 and neighbour not in closed_list:
                # Calculate the tentative g_score for reaching this neighbour:
                tentative_g_score = g_score[current_location] + 1
                # Check if the neighbour is either unvisited or has a lower g_score than previously recorded:
                if neighbour not in g_score or tentative_g_score < g_score[neighbour]:
                    # Record the current location as the previous location in `history`:
                    history[neighbour] = current_location

                    # Update the g_score for this neighbour:
                    g_score[neighbour] = tentative_g_score

                    # Calculate and update the f_score for this neighbour:
                    f_score[neighbour] = tentative_g_score + heuristic(neighbour, delivery_location)

                    # Push the neighbour and its f_score into the open list:
                    heapq.heappush(open_list, (f_score[neighbour], neighbour))

    #If no path is found, return None:
    return None
```

#### 4. Evaluation:
    * Let $b$ be the number of neighbouring nodes.
    * Let $d$ be the number of steps from the start to the goal.
##### Time Efficiency of A*:
    In the worst-case scenario, the time efficiency of A* is $O(b^d)$, meaning that every neighbouring node might be visited on the way from the start to the goal. This situation can arise with poor heuristic functions. Conversely, with a perfect heuristic function, A* will visit only that nodes on the path, resulting in a time efficiency of $O(d)$. In practice, the performance of A* typically falls between these two extremes and is heavily influenced by the accuracy of the heuristic function.
    $O(b^d)$
    Where:
    - $b$ is the branching factor, representing the average number of successors per state.
    - $d$ is the depth of the optimal solution, which in this case would correspond to the length of the shortest path from the start node to the goal node.
##### Space Efficiency of A*:
    Due to A* storing all neighbouring nodes as well as path nodes in memory, the space complexity in the worst case can be $O(b^d)$. This means that as the path length increases, the amount of memory used grows exponentially, making A* potentially unsuitable for large-scale pathfinding problems.
##### The Path Cost:
    The path cost of the solution is the total number of steps required to move from the starting location to the destination along the optimal path found by the algorithm. This is calculated as the length of the path minus one.

```python
# Function to output the path cost based on the results from the A* algorithm
def print_path_cost(results):
    # Check if the results are valid(i.e., not None or empty)
    if results:
        """
        results[0] : The path found by the A* algorithm
        results[1] : The current location of the drone after reaching the delivery location
        """
        # Extract the last location from the results to use for the next delivery (though not used in this function)
        last_location = results[1]
        # Store the path found in a variable
        path = results[0]
        # Print the path cost, which is the length of the path minus one
        # This represents the total number of steps taken (since the path includes the starting point)
        print(len(path) - 1)
    # Print an empty line for separation
    print()
```

```python
# Import the deepcopy feature to create a deep copy of the city map
import copy
# Path mapping function that takes two parameters: The city map and the path obtained from A*
def map_delivery_process(city_map, path):
    # Make a deep copy of the city map to preserve the original map
    # `deepcopy()` is necessary because the original `city_map` contains nested lists
    """
    `copy_of_city_map` is a duplicate of the original city map where the path will be marked with asterisks.
    """
    copy_of_city_map = copy.deepcopy(city_map)
    # Output the delivery number (assuming `index` is defined elsewhere)
    print(f"    *  Delivery map for delivery number {index + 1}:")
    """
    The path taken is marked with the "*" (asterisk) character
    The starting location is marked with the "S" character
    The destination is marked with the "E" character
    """
    # Iterate over each step in the path
    for step in path:
        # Mark each step in the path with an asterisk
        copy_of_city_map[step[0]][step[1]] = "*"
        # Check if the current location is the starting location
        if step == path[0]:
            # Mark the starting location with the letter "S"
            copy_of_city_map[step[0]][step[1]] = "S"
        # Check if the current location is the final location
        if step == path[-1]:
            # Mark the destination location with the letter "E"
            copy_of_city_map[step[0]][step[1]] = "E"
    # Return the modified copy of the map with the path visualized
    return copy_of_city_map
```

##### TKinter:
    TKinter is the standard GUI toolkit for Python. It is used to visualize the city map created in the previous function. In this implementation, a canvas is placed inside a window to display the city map. The map is represented with differently-colored tiles to illustrate various types of movements and path elements.

```python
#Import the `tkinter` package:
import tkinter
# Function to print the delivery process on a TKinter canvas
def print_delivery_process(canvas, city_map):
    # Get the number of rows in the city map
    rows = len(city_map)
    # Get the number of columns in the city map
    columns = len(city_map[0])
    # Set the size of each cell in the map (pixels):
    size = 40

    # Loop over each row in the city map
    for row in range(rows):
        # Loop over each column in the city map
        for column in range(columns):
            # Calculate the coordinates of the tile's corners
            corner_1 = column * size
            corner_2 = row * size
            corner_3 = corner_1 + size
            corner_4 = corner_2 + size

            # Initialize the default color for the letter on the tile
            letter_colour = ""
            # Determine the color of the tile based on its value
            if city_map[row][column] == 0:
                # Set the colours for passable tiles:
                colour = "black"
            # Check if the tile isn't passable:
            elif city_map[row][column] == 1:
                # Set the colours for unpassable tiles:
                colour = "white"
            # Check if the tile is the starting location:
            elif city_map[row][column] == "S":
                # Set the colour for the starting location's tile:
                colour = "cyan"
                # Set the colour for the letter on the tile:
                letter_colour = "yellow"
            #Check if the tile is the destination:
            elif city_map[row][column] == "E":
                # Set the colour for the destination's tile:
                colour = "yellow"
                # Set the colour for the letter on the tile:
                letter_colour = "cyan"
            #The tile can only be a path tile:
            else:
                #Set the colour for the path tile:
                colour = "darkgray"
            # Draw the tile on the canvas
            canvas.create_rectangle(corner_1, corner_2, corner_3, corner_4, fill = colour, outline = "green")
            # Place any text (letter) on the tile, if applicable
            canvas.create_text((corner_1 + corner_3) / 2, (corner_2 + corner_4) / 2, text=city_map[row][column], fill = letter_colour, font = ("Garamond", 20))
```

```python
# Loop over each city map and its corresponding delivery locations
for city_map, delivery_location in enumerate(delivery_locations):
    # Reset the last location to its origin for each new city map
    last_location = [0, 0]
    # Notify the user about the delivery process for the current city map
    print(f"•  The delivery process for city map {city_map + 1}:")
    # Loop through each delivery location in the current city map
    for index, delivery_location in enumerate(delivery_locations[city_map]):
        # Convert the last location to a tuple for use in the A* algorithm
        starting_location = tuple(last_location)
        # Calculate the path using the A* algorithm with the Chebyshev heuristic
        results = a_star(city_maps[city_map], starting_location, delivery_location, heuristicChebyshev)
        # Check if results were obtained from the A* algorithm
        if results:
            """
            results[0]: The path found by the A* algorithm
            results[1]: The final location of the drone after delivery
            results[2]: The length of the open list (count of nodes explored)
            results[3]: The length of the closed list (count of nodes visited)
            """
            # Update the starting location for the next delivery
            last_location = results[1]
            # Map the path on the city map
            current_map = map_delivery_process(city_maps[city_map], results[0])
            
            # Create a new Tkinter window
            window = tkinter.Tk()
            # Set the window title to include the delivery number and map number
            window.title(f"Delivery map for delivery {index + 1}, map {city_map + 1}")

            # Define the size and dimensions of the canvas
            rows = len(city_maps[city_map])
            columns = len(city_maps[city_map][0])
            size = 40
            canvas_width = columns * size
            canvas_height = rows * size

            # Create the canvas and place it in the window
            canvas = tkinter.Canvas(window, width=canvas_width, height=canvas_height)
            canvas.pack(expand=True)
            # Print the delivery process on the canvas
            print_delivery_process(canvas, current_map)

            # Start the TKinter main loop to display the window
            window.mainloop()
            """
            Tile colors:
            - Border color: "green"
            - Passable tiles: "black"
            - Non-passable tiles: "white"
            - Starting location: "cyan"
            - Destination: "yellow"
            - Path "darkgray"
            """

            #Output the length of the open list:
            print(f"\t=>  The count of neighbouring nodes explored: {len(results[2])}")
            #Output the length of the visited nodes:
            print(f"\t=>  The count of coordinates sets visited (excluding the final destination of this delivery): {len(results[3])}")
            #Let the user know that the next value to be printed out is the path cost:
            print(f"\t=>  The path cost for delivery {index + 1}:", end = " ")
            #Output the path cost for each delivery:
            print_path_cost(results)
        #In case no paths can be found:
        else:
            #Print out to the screen an error message for being unable find a path to a specific location, along with the delivery number:
            print(f"    *  There are no valid paths for delivery number {index + 1}.\n")
    #Output an empty line to separate the two test cases:
    print()
```
